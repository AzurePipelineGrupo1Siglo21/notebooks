```python
import pandas as pd
```

```python
prod_unico = pd.read_csv('Datasets locales/Producto_Unico.csv')
prod_unico.head()
```

```python
prod_suc = pd.read_csv('Datasets locales/Producto_Sucursales.csv')
prod_suc.head(10)
```

```python
prod_suc_collapsed = prod_suc.groupby('Cod_Producto').agg(lambda x: x.tolist())
prod_suc_collapsed.head()
```

```python
test = prod_unico.set_index('Cod_Producto').join(prod_suc_collapsed, on='Cod_Producto', how='inner')
test.head()
```

```python
for index, row in test.head().iterrows():
    print(f"Nombre: {row['Producto']}")
    for i in range(len(row['Cod_Sucursal'])):
        print(f"\tSucursal: {row['Cod_Sucursal'][i]} Stock: {row['Stock'][i]}")
```

```python
prueba_suc = prod_suc
prueba_suc['merged'] = prod_suc.apply(lambda row: {'Cod_Sucursal' : row['Cod_Sucursal'] , 'Stock' : row['Stock']}, axis=1)
prueba_suc.rename(columns={'merged':'Sucursal_y_Stock'}, inplace=True)
prueba_suc
```

```python
prueba_suc_collapsed = prueba_suc[['Cod_Producto', 'Sucursal_y_Stock']].groupby('Cod_Producto').agg(lambda x: x.tolist())
prueba_suc_collapsed.head()
```

```python
prueba_join = prod_unico.set_index('Cod_Producto').join(prueba_suc_collapsed, on='Cod_Producto', how='inner')
prueba_join.head()
```
